```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
usage=pd.read_csv('social_media_data.csv')
```

```python
usage.head()
```

```python
usage.shape
```

```python
usage['Category'].value_counts()
```

```python
usage['Category'].value_counts()[0:3]
```

```python
list(usage['Category'].value_counts()[0:3].keys())
```

```python
sns.histplot(data=usage, x='Likes', bins=30, kde=True)  
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Category', y='Likes', data=usage)
```

```python
likes_mean = usage['Likes'].mean()
print(f"The mean of the 'Likes' category is: {likes_mean}")
```

```python
by_category = usage.groupby('Category')['Likes'].mean()
print("Mean Likes for each Category:")
print(by_category)
```

```python
plt.figure(figsize=(8,4))
plt.bar(list(usage['Category'].value_counts()[0:8].keys()),list(usage['Views'].value_counts()[0:8].keys()),color="r")
```

```python
usage['Comments'].value_counts()
```

```python
plt.figure(figsize=(6,6))

plt.hist(usage['Followers Count'], bins=20, color='skyblue', alpha=0.7, label='Followers Count')

plt.hist(usage['Likes'], bins=20, color='green', alpha=0.7, label='Likes')

plt.title('Distribution of Followers Count and Likes')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Pie chart for 'Category'
plt.subplot(1, 3, 2)
plt.pie(usage['Category'].value_counts(), labels=None, autopct='%1.1f%%', colors=['orange', 'lightblue'])
plt.title('Category')

# Pie chart for 'Likes'
plt.subplot(1, 3, 3)
plt.pie(usage['Post Type'].value_counts(), labels=None, autopct='%1.1f%%', colors=['green', 'lightblue'])
plt.title('Post Type')
```
